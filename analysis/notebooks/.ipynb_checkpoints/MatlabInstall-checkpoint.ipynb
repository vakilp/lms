After I did the pip installs, I was having problem where the MATLAB kernet would come up and then die. The problem was that Jupyter could not find the matlab executable. So, I had to make the following changes in .bash_profile

export MATLAB_EXECUTABLE="/usr/local/bin/matlab"
export PATH="/Applications/anaconda/bin:$PATH"

I had to make a symbolic to the location of MATLAB to /usr/local/bin because adding the actual path from /Applicaitons did not seem to work. To add the symbolic link, this is what I did:

sudo ln -s /Applications/MATLAB_RXXXx.app/bin/matlab /usr/local/bin

```octave
t = 0:1/1280e6:1e-3;
t = t(1:end-1);
sig = awgn(repmat(sin(2*pi*200e6*t+pi*(25e6/1e-3)*t.^2),1,2),0,'measured');
size(sig)
```

```octave
length(sig)/4096
```

```octave
x = fft(reshape(sig,4096,625));
plot(db(abs(x)))
```
